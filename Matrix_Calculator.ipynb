```
class Matrix():

  def create_matrix1(self):
    self.row1, self.column1 = input("Enter size of first matrix: ").split()
    self.matrix1 = []
    self.matrixint = []
    print("Enter first matrix:")
    for i in range(int(self.row1)):
        self.matrix1.append(input().split())
    for k in range(len(self.matrix1)):
      self.matrixint.append([])
      for s in range(len(self.matrix1[0])):
        self.matrixint[k].append(float(self.matrix1[k][s]))
    return self.matrixint

  def create_matrix2(self):
    self.row2, self.column2 = input("Enter size of second matrix: ").split()
    self.matrix2 = []
    print("Enter second matrix:")
    for i in range(int(self.row2)):
        self.matrix2.append(input().split())
    return self.matrix2

  def addition(self):
    matrix1 = self.create_matrix1()
    matrix2 = self.create_matrix2()

    if self.row1 == self.row2 and self.column1 == self.column2:
      print("The result is:")
      for k in range(len(matrix1)):
        for s in range(len(matrix1[0])):
          print((float(matrix1[k][s]) + float(matrix2[k][s])), end=' ')
        print()
    else:
      print('Error! Matrices have different dimensions!')

  def matrix_by_constant(self):
    matrix = self.create_matrix1()
    x = int(input("Enter a constant: "))
    print("The result is:")
    for k in range(int(self.row1)):
      for s in range(int(self.column1)):
        print((int(matrix[k][s])) * x, end=' ')
      print()

  def matrix_by_matrix(self):
    matrix1 = self.create_matrix1()
    matrix2 = self.create_matrix2()
    result = [[0 for i in range(int(self.column2))] for j in range(int(self.row1))]

    if self.column1 == self.row2:
      print("The result is:")
      for k in range(len(matrix1)):
        for s in range(len(matrix2[0])):
          for r in range(len(matrix2)):
            result[k][s] += float(matrix1[k][r]) * float(matrix2[r][s])
      for k in range(len(result)):
        for s in range(len(result[k])):
          print(result[k][s], end=' ')
        print()
    else:
      print("Error! You cannot mulitply these matrices!")

  def transpose(self):
    print("1. Main diagonal")
    print("2. Side diagonal")
    print("3. Vertical line")
    print("4. Horizontal line")
    self.diagonal_choose = int(input())

    if self.diagonal_choose == 1:
      matrix = self.create_matrix1()
      transposed = []
      for k in range(len(matrix)):
        transposed.append([])
        for s in range(len(matrix[0])):
          transposed[k].append(int(matrix[s][k]))
      print("The result is:")
      for k in range(len(transposed)):
        for s in range(len(transposed[k])):
          print(transposed[k][s], end=' ')
        print()

    elif self.diagonal_choose == 2:
      matrix = self.create_matrix1()
      matrix_reversed = matrix[::-1]
      transposed = []
      for k in range(len(matrix_reversed)):
        transposed.append([])
        for s in range(len(matrix_reversed[0])):
          transposed[k].append(float(matrix_reversed[s][k]))
      transposed = transposed[::-1]
      print("The result is:")
      for k in range(len(transposed)):
        for s in range(len(transposed[k])):
          print(transposed[k][s], end=' ')
        print()

    elif self.diagonal_choose == 3:
      matrix = self.create_matrix1()

      mtx2 = []
      for k in range(len(matrix)):
        mtx2.append([])
        for s in range(len(matrix[0])):
          mtx2[k].append(float(matrix[s][k]))

      mtx2 = mtx2[::-1]
      transposed = []

      for k in range(len(mtx2)):
        transposed.append([])
        for s in range(len(mtx2[0])):
          transposed[k].append(int(mtx2[s][k]))
      print("The result is:")
      for k in range(len(transposed)):
        for s in range(len(transposed[k])):
          print(transposed[k][s], end=' ')
        print()

    elif self.diagonal_choose == 4:
      matrix = self.create_matrix1()
      transposed = matrix[::-1]
      print("The result is:")
      for k in range(len(transposed)):
        for s in range(len(transposed[k])):
          print(transposed[k][s], end=' ')
        print()
    else:
      print("That option does not exist!")

  def determinant(self, mat):
    det = 0
    if len(mat) == len(mat[0]):
      #print("The result is:")
      if len(mat) == 1:
        det = mat[0][0]
        return det
      elif len(mat) == 2:
        det = mat[0][0] * mat[1][1] - mat[0][1] * mat[1][0]
        return det
      else:
        for k in range(len(mat[0])):
          minor = [[mat[i][j] for j in range(len(mat[0])) if j != k] for i in range(1, len(mat))]
          #print(minor)
          det += self.determinant(minor) * mat[0][k] * (-1) ** k
          #print(det)
        return det
    else:
      return "Error! Matrix is not squared!"

exit = False
while exit == False:
    print("1. Add matrices")
    print("2. Multiply matrix by a constant")
    print("3. Multiply matrices")
    print("4. Transpose matrix")
    print("5. Calculate a determinant")
    print("0. Exit")

    test = Matrix()
    option = int(input())
    if option == 1:
      test.addition()
      print()
    elif option == 2:
      test.matrix_by_constant()
      print()
    elif option == 3:
      test.matrix_by_matrix()
      print()
    elif option == 4:
      test.transpose()
      print()
    elif option == 5:
      mat = test.create_matrix1()
      print("The result is:")
      print(test.determinant(mat))
      print()
    elif option == 0:
      exit = True
    else:
      print("Error!")



```
